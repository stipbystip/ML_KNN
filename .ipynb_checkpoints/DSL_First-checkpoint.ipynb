```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score
from sklearn.linear_model import LogisticRegression
```

```python
dataset = pd.read_csv('forest_dataset.csv')
dataset.head()
```

```python
dataset.info()
```

```python
#Информация о типе покрова содержится в последнем столбце
values_cover = dataset[dataset.columns[-1]].values
#Создание матрицы признаков
matrix_attributes = dataset[dataset.columns[:-1]].values
matrix_attributes[0]
```

```python
#Разделение датасета на Train и Test
matrix_attributes_Train, matrix_attributes_Test, values_cover_Train, values_cover_Test = train_test_split(
    matrix_attributes, values_cover, test_size = 0.2, random_state=42)
print(matrix_attributes_Train.shape, matrix_attributes_Test.shape, values_cover_Train.shape, values_cover_Test.shape)
```

```python

knn = 12
weights =['uniform', 'distance']
model = KNeighborsClassifier(n_neighbors=knn, weights='distance')
model.fit(matrix_attributes_Train, values_cover_Train)

prediction = model.predict(matrix_attributes_Test)
accuracy_score(values_cover_Test, prediction)
```
